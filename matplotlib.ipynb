# 2. Line Plot
* We use a line plot to show how something changes over time by connecting points with lines.
* It helps us see trends, patterns, or ups and downs in the data — like temperature over days, sales over months, or marks over exams.

```python
# We use a line plot to show how something changes over time by connecting points with lines.
years = [2006 + i for i in range(16)] #tatal 16 year data
print(years)

weight  = [80,83,48, 89,76,87,44,56,75,45,89,51,82,66,77,88]
print(weight)
```

```python
# Plot data and give it color we can use only first letter ling green = g
plt.plot(years, weight, c = 'green')
plt.show
```

```python
# Make line more thicker lw = line weidth
plt.plot(years,weight , c = 'g',lw = 3)
```

```python
# Change the linestyle plane to dot
plt.plot(years,weight , c = 'g',lw = 3,linestyle ="--" )
```

```python
# We can add title
plt.title('age_meter')

# We can add x and y axis name called label
plt.xlabel('Years')
plt.ylabel('Age')
plt.show
```

# 3. Bar Plot
* We use a bar plot to compare different categories or groups using rectangular bars.

```python
# Each bar shows the value of a category, and it's easy to see which is the biggest or smallest — like comparing sales of different products or marks in different subjects. 📊
# Data
subjects = ['Math', 'Science', 'English', 'History', 'Art']
marks = [90, 85, 78, 70, 88]
```

```python
# Create bar plot
plt.bar(subjects, marks, color='skyblue')
plt.title("Student Marks by Subject")
plt.xlabel("Subjects")
plt.ylabel("Marks")

# Set y-axis limit from 0 to 100
plt.ylim(0, 100) 
plt.show()
```

```python
# aline = wher a bar start 
plt.bar(subjects, marks, color='skyblue',align = "edge")

# set width
plt.bar(subjects, marks, color='skyblue',align = "edge", width = 0.5)

# Edge color 
plt.bar(subjects, marks, color='skyblue',align = "edge",edgecolor = "black")# lw for black width
```

# 5. Pie chart 
* A pie chart helps you see how much each part contributes to the total

```python
# Data
fruits = ['Apple', 'Banana', 'Mango', 'Orange', 'Grapes']
votes = [30, 25, 20, 15, 10]
```

```python
# Pie Chart
plt.pie(votes, labels=fruits)
```

```python
# Autopercentage autopct = 
##### Format	Output Example	Meaning
# '%1.1f%%'	45.0%	1 decimal place
# '%1.2f%%'	45.00%	2 decimal places
# '%1.3f%%'	45.000%	3 decimal places
# '%1.0f%%'	45%	No decimals (whole number)
plt.pie(votes, labels=fruits,autopct='%1.1f%%')


# Add some detail
plt.title("Favorite Fruits Survey")
plt.axis('equal')  # Makes it a perfect circle
plt.show()
```

```python
# Explodes = Use to pull out a part from chart .
# There is 5 value in chart
fruits = ['Apple', 'Banana', 'Mango', 'Orange', 'Grapes']
explodes = [0.1,0,0,0,0] # we can pull according to sequence
plt.pie(votes,labels = fruits, autopct = '%1.2f%%', explode = explodes)
plt.show()
```

```python
# In Matplotlib, the startangle parameter rotates the starting position of the first slice in a pie chart.
# Alwaus try to use 90
plt.pie(votes, labels = fruits,autopct = '%1.0f%%', startangle =45 )
plt.show()
```

* They help in reading values from the chart (like years, quantities, etc.)

* There are two types:

* yticks – ticks on the Y-axis

Parameter	Use

### Multiple Figure
 When to Use Multiple Figures?
Use it when:

* You want totally different plots (e.g., line chart and bar chart)

* You want to save each chart as a separate image

* You don’t want plots to overlap or mix

```python
# First figure
plt.figure(1)  # Or just plt.figure()
plt.plot([1, 2, 3], [4, 5, 6])
plt.title(" First Line Chart")

# Second figure
plt.figure(2)
plt.bar([1, 2, 3], [6, 5, 4])
plt.title(" Second Bar Chart")

plt.show()
```

###  SubPlots
 Why Use Subplots?
* To compare multiple charts side by side

* To save space in dashboards or reports

* To show related data together

```python
# plt.subplot(nrows, ncols, index)

# nrows	     Number of rows
# ncols 	Number of columns
# index 	Position of the current subplot
```

```python
import matplotlib.pyplot as plt

# Create a 1-row, 2-column layout
# plt.subplot(nrows, ncols, index)
plt.subplot(1, 2, 1)  # 1st plot in position 1
plt.plot([1, 2, 3], [4, 5, 6])
plt.title(" Line Chart")

plt.subplot(1, 2, 2)  # 2nd plot in position 2
plt.bar([1, 2, 3], [6, 5, 4])
plt.title(" Bar Chart")

plt.tight_layout()
plt.show()
```

### Exporting Plots
Exporting a plot means saving it as an image file (like .png, .jpg, .pdf, etc.) so you can:

*  Include it in a report

*  Share it with others

*  
*  Print or publish it



🔧 Exporting Options
You can customize the export using parameters:

* Option	Example	Meaning
* dpi	plt.savefig("plot.png", dpi=300)	High resolution (300 DPI for print) size of image
* transparent	transparent=True	No background color in PNG
* bbox_inches	'tight'	Crops out extra whitespace
* format	'pdf', 'svg', 'jpg', etc.	Set file format explicitly

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title(" Simple Line Chart")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.savefig("line_chart.png",dpi = 100,transparent = True)  # Saves the plot to your project folder
plt.show()
```

## 3D Ploting
* When Should You Use 3D Plots?
*  Use 3D plots only when you need to show relationships between three variables and when it adds real clarity to your data.
* What is 3D Plotting?
3D plotting lets you visualize three dimensions (X, Y, Z) in one chart — perfect for understanding patterns in 3D space, surfaces, or depth-related data.

```python
from mpl_toolkits.mplot3d import Axes3D  # Needed for 3D projection
import matplotlib.pyplot as plt
```

```python

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # 3D subplot

# Example data
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
z = [5, 15, 5, 10]

# Plotting a 3D line
ax.plot(x, y, z, color='green')

# Labels
ax.set_title(" 3D Line Plot")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")

plt.show()
```

# Important 

```python
#To enable 3D plotting, you must specify the projection='3d' parameter when creating a subplot.
ax = plt.axes(projection = "3d")
# there are 3 because of 3d 
x = [1,2,3,4]
y = [5,6,7,8]
z = [9,10,11,12]

ax.scatter(x,y,z)
ax.set_title("3D Plot")
plt.show()
```

```python
import numpy as np
#To enable 3D plotting, you must specify the projection='3d' parameter when creating a subplot.
ax = plt.axes(projection = "3d")
# there are 3 because of 3d 
x = np.arange(0,50,0.1)
y = np.sin(x)
z = np.cos(x)

ax.plot(x,y,z)
ax.set_title("3D Plot")
plt.show()
```

## Animating Plots
We use animation in Matplotlib to visually represent changes over time, explain complex processes step-by-step, and make data stories more engaging and easier to understand. 

```python
import random
```

```python
heads_tails = [0,0]
for _ in range(100000):
    heads_tails[random.randint(0,1)]+=1
    plt.bar(["Head","Tails"],heads_tails,color = ['red','blue'])
    plt.pause(0.001)
#plot.show()
```
